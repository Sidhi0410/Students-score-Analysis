#LETS FIND OUT THE IMPACT OF PARENTS EDUCATION  ON STUDENT SCORE #


```python
gb=df.groupby('ParentEduc').agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between parent education and student score")
```

```python

sns.heatmap(gb,annot=True)
plt.figure(figsize=(5,4))
plt.show
```

#Analysis we derive from this heat map is average score in math,reading and writing score is higher if parent has done masters and similarly
#its low if parents are high school passed! herby parent education is impacting more on eductaion.


```python
gb1=df.groupby('ParentMaritalStatus').agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between parent marital and student score")
plt.figure(figsize=(5,4))
plt.show
```

#From the above chart we have analysed that the marital status of parent has no impact or negligible impact on students score#
